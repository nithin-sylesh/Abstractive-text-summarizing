```python
def delete_links(input_text):
    pettern  = r'''(?i)\b((?:https?://|www\d{0,3}[.]|[a-z0-9.\-]+[.][a-z]{2,4}/)(?:[^\s()<>]+|\(([^\s()<>]+|(\([^\s()<>]+\)))*\))+(?:\(([^\s()<>]+|(\([^\s()<>]+\)))*\)|[^\s`!()\[\]{};:'",<>?«»“”‘’]))'''
    output = re.sub(pettern, ' ', input_text)
    output=re.sub(r"http\S+", "", output)
    return output
```

```python
def clean_text(input_text):
    replace = r'[/(){}\[\]|@âÂ,;\?\'\"\*…؟–’،!&\+-:؛-]'
    output = re.sub(replace, " ", input_text)
    words = nltk.word_tokenize(output)
    words = [word for word in words if word.isalpha()]
    output = ' '.join(words)
    return output
```

```python
def reducelengthening(input_text):
    reduce=r'(.)\1{2,}'
    output=re.sub(reduce,r"\1\1", input_text)
    return output
    
```

```python
def preprocessing(input_text):
    stopwords=[]
    tokenizer = nltk.tokenize.WhitespaceTokenizer() 
    tokens = tokenizer.tokenize(input_text)
    for i in tokens:
        if i not in text1:
            stopwords.append(i)
    pun_ = RegexpTokenizer(r"\w+")
    pnctuation_=pun_.tokenize(' '.join(stopwords))
    lemmatized_output = ' '.join([lemmatizer.lemmatize(w) for w in pnctuation_])
    return lemmatized_output
```

```python
def delete_stopwords(input_text):
    stop_words = set(nltk.corpus.stopwords.words("english"))
    tokenizer = nltk.tokenize.WhitespaceTokenizer()
    tokens = tokenizer.tokenize(input_text)
    wnl = nltk.WordNetLemmatizer()
    lemmatizedTokens =[wnl.lemmatize(t) for t in tokens]
    output = [w for w in lemmatizedTokens if not w in stop_words]
    output = ' '.join(output)
    return output
```
